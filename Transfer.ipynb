```python
import torchvision.models as models
from torch.optim import Adam
from torch.nn import BCELoss
import torch

from dataset_split import DatasetSplit
from model_utils import ModelUtils
from runner import Runner

%matplotlib inline
```

```python
small_dataset = DatasetSplit('./data/dogs_and_cats_small/')
dataloaders = small_dataset.get_dataloaders()
```

```python
model = models.vgg11(pretrained=True)
model.classifier = ModelUtils.get_binary_classifier()
```

```python
ModelUtils.freeze(model)
ModelUtils.unfreeze(model.classifier)
```

```python
optimizer = Adam(model.parameters())
loss = BCELoss()
runner = Runner(dataloaders, model, loss, optimizer)
```

```python
runner.run()
```

```python
runner.display_result()
```

```python
runner.save_result('small_transfer_result.pth')
```

```python
large_dataset = DatasetSplit('./data/dogs_and_cats_large/')
dataloaders = large_dataset.get_dataloaders()
```

```python
model = models.vgg11(pretrained=True)
model.classifier = ModelUtils.get_binary_classifier()
```

```python
ModelUtils.freeze(model)
for feature in range(16, 21):
    ModelUtils.unfreeze(model.features[feature])
ModelUtils.unfreeze(model.classifier)
```

```python
optimizer = Adam(
    [
        {'params': model.features.parameters(), 'lr': 0.0001},
        {'params': model.classifier.parameters()}
    ]
)
loss = BCELoss()
runner = Runner(dataloaders, model, loss, optimizer)
```

```python
runner.run()
```

```python
runner.display_result()
```

```python
runner.save_result('large_transfer_result.pth')
```
